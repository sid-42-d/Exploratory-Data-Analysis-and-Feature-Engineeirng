```python
##Combining the train and test datasets
final_df=pd.concat([train_df, test_df])
final_df.head()
```

```python
final_df.info()
```

```python
##Feature Engineering Process
##Handling the numerical Features
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.info()
```

```python
final_df.head()
```

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].astype(str).str.extract(r'(\d{2}:\d{2})')
```

```python
##Making the Arrival_Time a useful feature
final_df['Arrival_hour']=final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_minutes']=final_df['Arrival_Time'].str.split(':').str[1]
final_df.drop('Arrival_Time',axis=1,inplace=True)
final_df['Arrival_hour']=final_df['Arrival_hour'].astype(int)
final_df['Arrival_minutes']=final_df['Arrival_minutes'].astype(int)
```

```python
##Making the Departure Time a useful feature
final_df['Dep_hour']=final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_minutes']=final_df['Dep_Time'].str.split(':').str[1]
final_df.drop('Dep_Time',axis=1,inplace=True)
final_df['Dep_hour']=final_df['Dep_hour'].astype(int)
final_df['Dep_minutes']=final_df['Dep_minutes'].astype(int)
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, 'nan':1, '4 stops':4})
final_df.drop('Route',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df['Additional_Info'].unique()
```

```python
final_df['Duration_hour']=final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
final_df.drop(6474,axis=0,inplace=True)
final_df.drop(2660,axis=0,inplace=True)
```

```python
final_df['Duration_hour']=final_df['Duration_hour'].astype(int)
```

```python
final_df['Duration_hour']=final_df['Duration_hour']*60
final_df['Duration_hour'].astype(int)
```

```python
final_df['Duration_minutes']=final_df['Duration'].str.split('').str[1].str.split('m').str[0]
```

```python
final_df['Duration_minutes']=final_df['Duration_minutes'].astype(int)
```

```python
final_df['TotalDuration_Min']=final_df['Duration_minutes']+final_df['Duration_hour']
```

```python
final_df.drop(['Duration','Duration_minutes','Duration_hour'],axis=1,inplace=True)
```

```python
final_df['Airline'].unique()
```

```python
##Handling the categorical Features
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
final_df['Airline']=labelencoder.fit_transform(final_df['Airline'])
final_df['Source']=labelencoder.fit_transform(final_df['Source'])
final_df['Destination']=labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info']=labelencoder.fit_transform(final_df['Additional_Info'])
```
